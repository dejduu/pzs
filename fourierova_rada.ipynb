```python
import numpy as np
import matplotlib.pyplot as plt

# Definujeme funkci a poet len Fourierovy ady
def fourier_series_approximation(t, N):
    sum_series = np.zeros_like(t)
    for n in range(1, N + 1):
        bn = -2 / n * np.cos(n * np.pi)
        sum_series += bn * np.sin(n * t)
    return sum_series

# Rozsah pro t od -pi do pi
t = np.linspace(-np.pi, np.pi, 1000)
f_t = t  # Funkce f(t) = t

# Poet len Fourierovy ady, kter chceme pout
N_values = [1,2,200]

# Vykreslme grafy pro rzn hodnoty N
plt.figure(figsize=(12, 8))
for N in N_values:
    f_approx = fourier_series_approximation(t, N)
    plt.plot(t, f_approx, label=f'N = {N}')

# Pvodn funkce f(t) = t
plt.plot(t, f_t, 'k--', label='f(t) = t')

plt.xlabel('t')
plt.ylabel('f(t)')
plt.title('Aproximace f(t) = t pomoc Fourierovy ady')
plt.legend()
plt.grid(True)
plt.show()
```
