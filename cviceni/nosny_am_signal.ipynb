```python
import numpy as np
import matplotlib.pyplot as plt

def nosny_signal(t, A0, omega):
    return A0 * np.sin(omega * t)

def modulacni_signal(t, M, phi, E):
    return M * np.sin(phi * t + E)

def am_signal(t, A0, omega, m_t):
    return (A0 + m_t) * np.sin(omega * t)

# Parametry
A0 = 1.0
omega = 2 * np.pi * 5
t = np.linspace(0, 1, 1000)

M = 0.5
phi = 2 * np.pi * 2
E = np.pi / 4

# Vpoty signl
f_t = nosny_signal(t, A0, omega)
m_t = modulacni_signal(t, M, phi, E)
s_am = am_signal(t, A0, omega, m_t)

plt.figure(figsize=(12, 6))
plt.plot(t, f_t, label=r'Nosn signl $f(t)$')
plt.plot(t, s_am, label=r'AM signl $s_{AM}(t)$', color='orange')
plt.title('Nosn a AM signl')
plt.xlabel('as [s]')
plt.ylabel('Amplituda')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

minimum = np.min(s_am)
maximum = np.max(s_am)

vykon = np.mean(s_am**2)

dt = t[1] - t[0]
energie = np.sum(s_am**2) * dt

stredni_hodnota = np.mean(s_am)
#zaokrouhleno = round(stredni_hodnota, 15)
rozptyl = np.var(s_am)

print("prumer: ", stredni_hodnota )
#print("pruemr:", zaokrouhleno)
print("Rozptyl:", rozptyl)

print("Minimum:", minimum)
print("maximum:", maximum)
print("Vykon:", vykon)
print("Energie: ", energie)

window_size = 50
trend = np.convolve(s_am, np.ones(window_size)/window_size, mode='same')
#proste prumer hodnot za nejaky obdobi

plt.figure(figsize=(12, 4))
plt.plot(t, s_am, label='AM signl')
plt.plot(t, trend, label='Trend (pom. prmru)', color='red', linewidth=1)
plt.legend()
plt.title('AM signl a jeho trend')
plt.xlabel('as [s]')
plt.ylabel('Amplituda')
plt.grid(True)
plt.tight_layout()
plt.show()
```
