```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
sampling_rate = 1000  # in Hz
duration = 1.0        # in seconds
frequency = 5         # signal frequency in Hz
noise_level = 1    # adjust this value to increase or decrease noise

# Generate time vector
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Generate clean sine wave signal
signal = np.sin(2 * np.pi * frequency * t)

# Generate noise (Gaussian noise)
noise = noise_level * np.random.normal(size=t.shape)

# Add noise to the signal
noisy_signal = signal + noise

# Define the window size for moving average (shadowing)
window_size = 20  # Adjust this size based on desired smoothing effect

# Apply moving average filter (shadowing)
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'same') / w

# Get the shadowed (smoothed) signal
shadowed_signal = moving_average(noisy_signal, window_size)

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(t, noisy_signal, label='Noisy Signal', alpha=0.5)
plt.plot(t, shadowed_signal, label='Shadowed (Smoothed) Signal', linewidth=2)
plt.plot(t, signal, label='Original Clean Signal', linestyle='--', alpha=0.8)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.title('Signal Shadowing with Moving Average Filter')
plt.legend()
plt.show()
```
