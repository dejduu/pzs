```python
import numpy as np
import matplotlib.pyplot as plt

# Parametry signlu
f = 5  # Hz
fs = 1000  # vzorkovn
d = 2  # dlka v s

# Jeden asov vektor pro vechny signly
t = np.linspace(0, d, int(fs * d), endpoint=False)

# Sinusov signl
y_sin = np.sin(2 * np.pi * f * t)

# Pilov signl z asovho vektoru (nepmo ze sinusu, ale stle z t)
y_pil = np.zeros_like(y_sin)
for i in range(len(t)):
    y_pil[i] = 2 * ((f * t[i]) % 1) - 1

# Trojhelnkov signl z asovho vektoru
y_troj = np.zeros_like(y_sin)
for i in range(len(t)):
    y_troj[i] = 2 * np.abs(2 * ((f * t[i]) % 1) - 1) - 1

# Obdlnkov signl ze sinusoidy
y_obd = np.sign(y_sin)

# Impulsov signl z asovho vektoru
y_imp = np.zeros_like(y_sin)
perioda = int(fs / f)
y_imp[::perioda] = 1

# Zobrazen vech signl
plt.figure(figsize=(10, 10))
plt.subplot(5, 1, 1)
plt.plot(t, y_sin)
plt.title('Sinusov signl')

plt.subplot(5, 1, 2)
plt.plot(t, y_pil)
plt.title('Pilov signl')

plt.subplot(5, 1, 3)
plt.plot(t, y_troj)
plt.title('Trojhelnkov signl')

plt.subplot(5, 1, 4)
plt.plot(t, y_obd)
plt.title('Obdlnkov signl')

plt.subplot(5, 1, 5)
plt.stem(t, y_imp)
plt.title('Impulsov signl')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sinusovy_signal(frekvence, vzorkovaci_frekvence, delka, amplituda=1, faze=0):
    t = np.linspace(0, delka, int(vzorkovaci_frekvence * delka), endpoint=False)
    return t, amplituda * np.sin(2 * np.pi * frekvence * t + faze)

def pilovy_signal(frekvence, vzorkovaci_frekvence, delka, amplituda=1):
    t = np.linspace(0, delka, int(vzorkovaci_frekvence * delka), endpoint=False)
    #  sin
    y = amplituda * (2 * ( (frekvence * t) % 1 ) - 1)
    return t, y

def trojuhelnikovy_signal(frekvence, vzorkovaci_frekvence, delka, amplituda=1):
    t = np.linspace(0, delka, int(vzorkovaci_frekvence * delka), endpoint=False)
    # pomoc abs(sin)
    y = amplituda * (2 * np.abs(2 * ((frekvence * t) % 1) - 1) - 1)
    return t, y

def obdelnikovy_signal(frekvence, vzorkovaci_frekvence, delka, amplituda=1):
    t = np.linspace(0, delka, int(vzorkovaci_frekvence * delka), endpoint=False)
    # sign(sin)
    y = amplituda * np.sign(np.sin(2 * np.pi * frekvence * t))
    return t, y

def impulsovy_signal(frekvence, vzorkovaci_frekvence, delka, amplituda=1):
    t = np.linspace(0, delka, int(vzorkovaci_frekvence * delka), endpoint=False)
    y = np.zeros_like(t)
    perioda = int(vzorkovaci_frekvence / frekvence)
    y[::perioda] = amplituda
    return t, y

# Parametry signlu
f = 5 #hz
fs = 1000 #vzorkovani
d = 2 #delka v s

# Generovn signl
t, y_sin = sinusovy_signal(f, fs, d)
t_pil, y_pil = pilovy_signal(f, fs, d)
t_troj, y_troj = trojuhelnikovy_signal(f, fs, d)
t_obd, y_obd = obdelnikovy_signal(f, fs, d)
t_imp, y_imp = impulsovy_signal(f, fs, d)

plt.figure(figsize=(10, 10))
plt.subplot(5, 1, 1)
plt.plot(t, y_sin)
plt.title('Sinusov signl')

plt.subplot(5, 1, 2)
plt.plot(t_pil, y_pil)
plt.title('Pilov signl')

plt.subplot(5, 1, 3)
plt.plot(t_troj, y_troj)
plt.title('Trojhelnkov signl')

plt.subplot(5, 1, 4)
plt.plot(t_obd, y_obd)
plt.title('Obdlnkov signl')

plt.subplot(5, 1, 5)
plt.stem(t_imp, y_imp)
plt.title('Impulsov signl')

plt.tight_layout()
plt.show()
```
